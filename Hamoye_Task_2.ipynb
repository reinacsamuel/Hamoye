### Seperating Numeric data from Categorical data

```python
# numeric data
data_numeric = data.select_dtypes(include=[np.number])
numeric_cols = data_numeric.columns.values
print(f'There are {len(numeric_cols)} numeric columns in the dataset','\n')
print(numeric_cols)
```

```python
# categorical data
data_categorical = data.select_dtypes(exclude=[np.number])
non_numeric_cols = data_categorical.columns.values
print(f'There are {len(non_numeric_cols)} non-numeric columns in the dataset','\n')
print(non_numeric_cols)
```

```python
#SPOTTING MISSING VALUES
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
# Visualizing missing data using heatmap

cols = data.columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(data[cols].isnull(), cmap=sns.color_palette(colours))
```

#### From the above heatmap, we see the missing values in yellow color

```python
#PRESENTING MISSING VALUES IN PERCENTAGE %
```

```python
# if it's a larger dataset and the visualization takes too long can do this.
# % of missing.
for col in data.columns:
#     missing_num = data[col].isnull().sum()
    pct_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100,4)))
```

```python
data.shape
```

```python
#Imputing the missing values of individual columns by their respective mean values
```

```python
for col in numeric_cols:
    data[col] = data[col].fillna(data[col].mean())
```

```python
data[numeric_cols].isnull().sum()
```

```python
data.shape
```

```python
#After filling the missing values, the yellow colors are gone
```

```python
# Visualizing missing data using heatmap

cols = data.columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(data[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
data.isnull().sum()
```

```python
#From the above, we realize that the first four columns still have missing values because they are objects and arithmetic mean isn't applicable. Let's use mode to fill them up
```

```python
print(non_numeric_cols)
```

```python
data[non_numeric_cols].isnull().sum()
```

```python
#Filling individual categorical column with the most frequent column values
```

```python
for non_numeric_col in non_numeric_cols:
    data[non_numeric_col] = data[non_numeric_col].fillna(data[non_numeric_col].value_counts().index[0])
```

```python
data[non_numeric_cols].isnull().sum()
```
